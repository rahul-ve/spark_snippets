```
import datetime as dt
import jaydebeapi
from pyspark.sql.functions import *
```

### Using JayDeBeApi module as JDBC client to interact with SQL servers, similar to pyodbc but uses the bundled jdbc drivers!

```
#SQL server config

jdbcHostname = "xxxxxx.database.windows.net" 
jdbcDatabase = "xxxx"                                            
jdbcPort = 1433
jdbcUsername = dbutils.secrets.get(scope = "azure-kv", key = "db-un")
jdbcPassword = dbutils.secrets.get(scope = "azure-kv", key = "db-pw")
MSSQLdriver = "com.microsoft.sqlserver.jdbc.SQLServerDriver"
jdbcUrl = "jdbc:sqlserver://{0}:{1};database={2}".format(jdbcHostname, jdbcPort, jdbcDatabase)
connectionProperties = {
  "user" : jdbcUsername,
  "password" : jdbcPassword,
  "driver" : MSSQLdriver
}
```

```
jdbcUrl_str = "jdbc:sqlserver://{host}:{port};database={db_elem};user={user};password={password}".format(
                                                      host=jdbcHostname, port=jdbcPort, db_elem=jdbcDatabase,
                                                      user=jdbcUsername, password=jdbcPassword)
conn = jaydebeapi.connect(MSSQLdriver, jdbcUrl_str)
curs = conn.cursor()
```

```
utc_isotime = dt.datetime.now(dt.timezone.utc).strftime('%Y-%m-%d %H:%M:%S')
```

```
query = """UPDATE dbo.testing_times
SET 
EventType = 'ET',
ModifiedOn = '{event_time}'
WHERE EventKey = 'K1'
""".format(event_time = utc_isotime)
```

```
curs.execute(query)
```

```
curs.close()
conn.close()
```
